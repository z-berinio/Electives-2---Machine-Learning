```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline    
import seaborn as sns

```

#Humidity

Read the dataset

```python
from google.colab import files
data_to_load = files.upload()


 
```

```python
import io
df = pd.read_csv(io.BytesIO(data_to_load['humidity.csv']))
```

Get the descriptive statistics

```python
df.describe()
```

 Visualize humidity 

```python
df.plot(kind="scatter", x ="sample", y= 'humidity', alpha= .4, 
            s=df["humidity"]/100, label="humidity", figsize=(20,8),
            c="sample", cmap=plt.get_cmap("jet"), colorbar=True,
            )
```

Plot the attributes distribution (humidity)

```python
sns.distplot(df['sample'],50)
plt.show()
```

```python
sns.distplot(df['humidity'],50)
plt.show()
```
